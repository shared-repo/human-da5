```python
# df_population2['행정구역'].str.slice(-6, -5) # '행정구역' 컬럼에서 뒤에서 여섯번째 문자 읽기
# df_population2['행정구역'].str.slice(-6, -5) != '0' # '행정구역' 컬럼에서 뒤에서 여섯번째 문자가 '0'이 아닌지 여부
df_population2["읍면동여부"] = df_population2['행정구역'].str.slice(-6, -5) != '0'
df_population2.iloc[:5, [0, -1]]
```

```python
# 읍면동 수준에서 인구가 가장 많은, 가장 적은 행정구역, 인구가 가장 적은 행정구역 검색
mask = df_population2['읍면동여부']
max_pop = df_population2[mask]['2025년01월_계_총인구수'].max()
min_pop = df_population2[mask].iloc[:, 1].min()
print( max_pop, min_pop )

col = df_population2['2025년01월_계_총인구수']
# (col == max_pop) or (col == min_pop) # 오류 : and, or는 여러 개의 값에 대한 논리 연산을 한 번에 실행 불가
mask2 = (col == max_pop) | (col == min_pop) # 총인구수 컬럼의 값이 max_pop 또는 min_pop와 같은 행 조회
df_population2[mask & mask2].iloc[:, :2]
```

```python
# 남성인구비가 가장 높은 행정동, 여성인구비가 가장 높은 행정동

# 컬럼 확인
print( df_population2.columns[[1, 104, 207]] )

# 남성인구비 컬럼 만들기
df_population2['남성인구비'] = df_population2.iloc[:,104] / df_population2.iloc[:,1]
df_population2['여성인구비'] = df_population2['2025년01월_여_총인구수'] / df_population2['2025년01월_계_총인구수']
df_population2.iloc[:5, [0, -2, -1]]
mask1 = df_population2['읍면동여부']
mask2 = df_population2['남성인구비'] == df_population2['남성인구비'].max()
mask3 = df_population2['여성인구비'] == df_population2['여성인구비'].max()

print( df_population2[mask1 & mask2].iloc[:, [0, -2, -1]] )
print( df_population2[mask1 & mask3].iloc[:, [0, -2, -1]] )
```

```python
# 여성인구비가 가장 높은 행정동의 연령대별 인구 분포 시각화
print( df_population2.columns[[209, 309]] ) # 여성 연령별 인구 데이터 컬럼 확인

df_population2[mask3].iloc[:, 209:310] # 여성인구비가 가장 높은 행정동의 연령대별 인구 데이터 조회
ar_pop = df_population2[mask3].iloc[:, 209:310].values # values : dataframe or series --> ndarray
print( ar_pop, ar_pop.shape )
# ar_pop2 = ar_pop.reshape(101) # shape을 101개 요소를 갖는 1차원 배열로 변환
# ar_pop2 = ar_pop.reshape(-1) # shape을 1차원 배열로 변환 (요소 갯수는 자동 계산)
ar_pop2 = ar_pop.ravel() # shape을 1차원 배열로 변환 (요소 갯수는 자동 계산)
print(ar_pop2, ar_pop2.shape)

plt.bar(np.arange(len(ar_pop2)), ar_pop2)
plt.show()
```

```python
# 여성인구비가 중간인 행정동 찾기 --> 비교대상을 서울시로 선택
mid_value = df_population2['여성인구비'].median() # 중앙값
# mask = df_population2['여성인구비'] == mid_value
diff = (df_population2['여성인구비'] - mid_value)
diff = diff.abs()
mask = diff < 0.05
df_population2[mask].iloc[:, [0, -1]]
```

```python
# 서울시 연령별 인구분포 시각화
df_population2.iloc[0, 209:310] # 서울시 연령별 여성 인구 데이터
seoul_female_pop = df_population2.iloc[0, 209:310].values.flatten() # 서울시 연령별 여성 인구 데이터를 1차원 배열로 만들기

plt.bar(range(len(seoul_female_pop)), seoul_female_pop)
plt.show()
```
