```python
# 읍면동 수준에서 인구가 가장 많은, 가장 적은 행정구역, 인구가 가장 적은 행정구역 검색
mask = df_population2['읍면동여부']
max_pop = df_population2[mask]['2025년01월_계_총인구수'].max()
min_pop = df_population2[mask].iloc[:, 1].min()
print( max_pop, min_pop )

col = df_population2['2025년01월_계_총인구수']
# (col == max_pop) or (col == min_pop) # 오류 : and, or는 여러 개의 값에 대한 논리 연산을 한 번에 실행 불가
mask2 = (col == max_pop) | (col == min_pop) # 총인구수 컬럼의 값이 max_pop 또는 min_pop와 같은 행 조회
df_population2[mask2].iloc[:, :2]
```
