```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# pandas의 read_excel 실행을 위해 필요한 패키지 설치
# !pip install openpyxl
```

```python
df_cctv = pd.read_excel('data-files/cctv-in-seoul.xlsx', header=2)
# df_cctv2 = df_cctv.drop(["Unnamed: 0", "순번"], axis=1) #지정된 이름의 열 삭제 (원본 변경 X)
# df_cctv2.head()

# df_cctv.drop(["Unnamed: 0", "순번"], axis=1, inplace=True) # inplace=True : 원본을 변경하는 설정
df_cctv.drop([df_cctv.columns[0], "순번"], axis=1, inplace=True) # inplace=True : 원본을 변경하는 설정
df_cctv.drop([0,26], axis=0, inplace=True)
# df_cctv.reset_index() # 기존 index를 일반 컬럼으로 옮기고 새 index 생성
df_cctv.reset_index(drop=True, inplace=True) # 기존 index를 제거하고 새 index 생성
# df_cctv.rename(columns={"2015년 이전 설치된 CCTV ":"2015년", "구분":"자치구"}, inplace=True)
df_cctv.rename(columns={df_cctv.columns[2]:"2015년", df_cctv.columns[0]:"자치구"}, inplace=True)
df_cctv
```

```python
# cctv가 제일 많은  / 제일 적은 자치구 확인
mask = df_cctv['총 계'] == df_cctv['총 계'].max()
print( df_cctv[mask] )

# df_cctv.sort_values("총 계").iloc[[0,-1],:] # sort_values : 지정된 컬럼 기준 정렬 수행 (기본 오름차순 정률)
df_cctv.sort_values("총 계", ascending=False).iloc[[0,-1],:] # ascending : 정렬기준 결정 (오름차순/내림차순)
```

```python
# 최근 3년 증가율 컬럼 만들기 : 최근 3년 신규 설치 합 / 그 이전 신규 설치 합
df_cctv['최근증가율'] = df_cctv.iloc[:,9:12].sum(axis=1) / df_cctv.iloc[:,2:9].sum(axis=1)
df_cctv.head()

# 최근 3년 증가율이 높은 자치구 확인
df_cctv.sort_values(by="최근증가율", ascending=False)
```
