```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_population = pd.read_excel('data-files/population-in-seoul.xlsx', header=1)
# df_population.drop(df_population.columns[0], axis=1).drop(0, axis=0) # (합계 컬럼, 총계 행 제거)
df_population = df_population.iloc[1:, 1:] # drop과 같은 효과 발생 (합계 컬럼, 총계 행 제거)
df_population = df_population.drop(df_population.columns[1:5], axis=1) #  (2024 년이 중복되어서 제거)

new_columns = ["자치구"] # 첫 번째 컬럼명
for year in np.arange(2014, 2025):
    new_columns.append(f"{year}-인구수")
    new_columns.append(f"{year}-한국인")
    new_columns.append(f"{year}-외국인")
    new_columns.append(f"{year}-고령자")
    # new_columns += [f"{year}-{c}" for c in ['인구수', '한국인', '외국인', '고령자']]

df_population.columns = new_columns # 전체 컬럼명을 모두 변경

df_population.reset_index(drop=True, inplace=True)
df_population
```

```python
# 2024년도 데이터만 추출
df_population_2024 = df_population.iloc[:, [0,-4,-3,-2,-1]].copy()
df_population_2024.head(3)
```

```python
# 2024년 데이터에 외국인 비율 컬럼과 고령자 비율 컬럼 추가
df_population_2024['외국인비율'] = df_population_2024['2024-외국인'] / df_population_2024['2024-인구수']
df_population_2024['고령자비율'] = df_population_2024['2024-고령자'] / df_population_2024['2024-인구수']
df_population_2024.head()
```

```python
# 외국인 비율 높은 자치구 / 고령자 비율 높은 자치구 조회
df_population_2024.sort_values(by="고령자비율", ascending=False)
df_population_2024.sort_values(by="외국인비율", ascending=False)
```

```python
df_population.to_csv('data-files/population-in-seoul-2.csv', index=False)
df_population_2024.to_csv('data-files/population-in-seoul-2024.csv', index=False)
```
