```python
from urllib import request
import requests
from bs4 import BeautifulSoup
```

```python
resp = requests.get("https://news.naver.com/breakingnews/section/105/230")
html = resp.content.decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```

```python
news_list = []
# soup.select("ul > li")
li_list = soup.select("ul.sa_list > li.sa_item")
for li in li_list:
    # print(li)
    a = li.select_one("div.sa_text a")
    div = li.select_one("div.sa_text div.sa_text_lede")
    # print(a.text, div.text)
    news_list.append((a.text.replace("\n", ""), div.text.replace("\n", "")))

print(news_list)
```

```python
import pandas as pd
```

```python
import time

titles = []
urls= []
descs = []
articles = []
# soup.select("ul > li")
li_list = soup.select("ul.sa_list > li.sa_item")
for li in li_list:
    # print(li)
    a = li.select_one("div.sa_text a")
    div = li.select_one("div.sa_text div.sa_text_lede")
    titles.append(a.text.replace("\n", ""))
    urls.append(a.get("href"))
    descs.append(div.text.replace("\n", ""))

for idx, url in enumerate(urls):
    print(f"{idx + 1}번째 뉴스 처리중")
    resp = requests.get(url)
    html = resp.content.decode('utf-8')
    soup2 = BeautifulSoup(html, "html.parser")
    article = soup2.select_one('article#dic_area')
    articles.append(article.get_text())
    time.sleep(1)

df_news = pd.DataFrame({"title": titles, "urls": urls, "desc": descs, "article": articles})
```

```python
df_news.to_csv("data-files/news.csv", encoding="utf-8", index=False)
```
