```python
from urllib import request
import requests
from bs4 import BeautifulSoup

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
header_info = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36"
}

resp = requests.get('https://datalab.naver.com/', headers=header_info)
html = resp.content.decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.prettify())
```

```python
# print( len( soup.select("div.keyword_rank") ) )
rows = []
divs = soup.select("div.keyword_rank")
for div in divs:
    row = []
    title = div.select_one('strong.rank_title').get_text().replace("\n", "").strip()
    row.append(title)
    # lis = div.select('ul.rank_list li.list')
    lis = div.select('li')
    for li in lis:
        num = li.select_one('.num').get_text().replace("\n", "").strip()        
        keyword = li.select_one('.title').get_text().replace("\n", "").strip()
        row.append( f"{num}. {keyword}" )
    rows.append(row)
```
