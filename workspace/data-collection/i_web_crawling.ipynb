```python
from urllib import request
import requests
from bs4 import BeautifulSoup

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# get 요청은 requests.get 함수 사용 / post 요청은 requests.post 함수 사용
resp = requests.post(url="https://dhlottery.co.kr/gameResult.do?method=byWin",
                     data={ "drwNo": 1147, "hdrwComb": 1, "dwrNoList": 1147 })
```

```python
html = resp.content.decode("euc-kr").replace("\n", "").replace("\r", "")
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.prettify())
```

```python
spans = soup.select(".win_result span")
for span in spans:
    print(span.get_text())
```
