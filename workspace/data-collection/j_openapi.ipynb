```python
client_id = "A7EXn1cdgHCSGjuWokoF"
client_secret = "6Fb0bI3OtH"
url = "https://openapi.naver.com/v1/search/book.json" # JSON 결과

headers = {
    "X-Naver-Client-Id": client_id,
    "X-Naver-Client-Secret": client_secret
}
params = {
    "query": "챗봇",
    "diaplay": 100,
    "start": 1,
    "sort": "date"
}

response = requests.get(url, params=params, headers=headers)
rescode = response.status_code # 응답 코드 읽기
if(rescode==200): # 200 : 정상응답
    response_body = response.content # 응답 본문 읽기
    # print(response_body.decode('utf-8'))
    json_str = response_body.decode('utf-8')
else:
    print("Error Code:" + rescode)
```
