```python
import csv
```

```python
rows = []
with open('data-files/subway-data-2.csv', 'rt', encoding='utf-8') as f:
    reader = csv.reader(f) # csv 문법 규칙을 반영해서 데이터를 읽는 도구

    for i, line in enumerate(reader): # 지정된 파일을 csv 문법 규칙에 따라 한 줄씩 읽어서 리스트로 반환
        rows.append(line[:4] + \
                    [ int(v.replace(",", "")) for v in line[4:-1] ] )
```

```python
print( rows[0] )
```

```python
# 출근시간(07시 ~ 09시) 이용객수가 가장 많은 역

high = 0
high_row = None
for row in rows:
    t = sum(row[10:14]) # 07 ~ 09 이용객수
    if high < t:
        high = t
        high_row = row

print(high_row[1], high_row[3], high)
```

```python
# 퇴근시간(17시 ~ 20시) 이용객수가 가장 많은 역

high = 0
high_row = None
for row in rows:
    t = sum(row[30:36]) # 17 ~ 20 이용객수
    if high < t:
        high = t
        high_row = row

print(high_row[1], high_row[3], high)
```

```python
# 전체 이용객수가 가장 많은 역

high = 0
high_row = None
for row in rows:
    t = sum(row[4:]) # 17 ~ 20 이용객수
    if high < t:
        high = t
        high_row = row

print(high_row[1], high_row[3], high)
```
